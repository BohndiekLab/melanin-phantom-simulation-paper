```python
import matplotlib.pyplot as plt
from PIL import Image

from patato.useful_utilities import add_axis_label, add_subfigure_label
import matplotlib
import numpy as np
import scienceplots

plt.style.use(["science", "nature", "no-latex"])

matplotlib.rcParams["font.sans-serif"] = ["Arial"]
matplotlib.rcParams["figure.constrained_layout.use"] = True

from patato.useful_utilities import add_axis_label, add_subfigure_label
```

```python
height = 255
width = 96
im1 = Image.open("../data/mice/mouse photos/notpigmented_back.png")
im2 = Image.open("../data/mice/mouse photos/notpigmented_front.png")
im3 = Image.open("../data/mice/mouse photos/pigmented_back.png")
im4 = Image.open("../data/mice/mouse photos/pigmented_front.png")


def crop_to_size(im1):
    im1 = im1.crop(
        (
            im1.size[0] // 2 - width // 2,
            im1.size[1] // 2 - height // 2,
            im1.size[0] // 2 + width // 2,
            im1.size[1] // 2 + height // 2,
        )
    )
    return im1


im1 = crop_to_size(im1)
im2 = crop_to_size(im2)
im3 = crop_to_size(im3)
im4 = crop_to_size(im4)
```

```python
fig = plt.figure(figsize=(5.25, 5 * 5.25 / 8))

subfig1, subfig2 = fig.subfigures(1, 2)

axes1 = subfig1.subplots(1, 2)
axes2 = subfig2.subplots(1, 2)

for ax in axes1:
    ax.axis("off")
for ax in axes2:
    ax.axis("off")

axes1[0].imshow(im1)
axes1[1].imshow(im2)
axes2[0].imshow(im3)
axes2[1].imshow(im4)

# subfig1.suptitle("Non-pigmented B6")
# subfig2.suptitle("Pigmented B6")

axes1[0].set_title("Back")
axes1[1].set_title("Front")
axes2[0].set_title("Back")
axes2[1].set_title("Front")

subfig1.suptitle("Non-pigmented")
subfig2.suptitle("Pigmented")

# axes1[0].annotate("Non-pigmented", (0.5, 0.1), xycoords="subfigure fraction", fontweight="bold", fontsize="large", ha="center")
# axes2[0].annotate("Pigmented", (0.5, 0.1), xycoords="subfigure fraction", fontweight="bold", fontsize="large", ha="center")

add_subfigure_label(subfig1, axes1[0], "A")
add_subfigure_label(subfig2, axes2[0], "B")
fig.savefig("../figures/mouse_pigmentation.png", dpi=300)
fig.savefig("../figures/mouse_pigmentation.pdf", dpi=300)
plt.show()
```
